```python
import matplotlib
matplotlib.use('TkAgg')  # O cualquier otro backend interactivo que funcione en tu sistema

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
import matplotlib.pyplot as plt

# Creaci√≥n de registros
q = QuantumRegister(1, 'q_reg')
c = ClassicalRegister(1, 'c_reg')

# Creaci√≥n del circuito
qc = QuantumCircuit(q, c)
qc.s(q[0])
qc.measure(q[0], c[0])

# Dibujando el circuito
fig = qc.draw(output='mpl')
plt.show()
```

# Aplicando "X" y job.result

```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, Aer, execute
from qiskit.visualization import plot_bloch_multivector

# Creaci√≥n de registros
q = QuantumRegister(1, 'q_reg')
c = ClassicalRegister(1, 'c_reg')

# Creaci√≥n del circuito
qc = QuantumCircuit(q, c)
qc.x(q[0])  # Aplicar la puerta X al qubit q
qc.measure(q[0], c[0])

# Dibujando el circuito
fig = qc.draw(output='mpl')
plt.show()

# Simulador
sim = Aer.get_backend('statevector_simulator')

# Ejecutar el circuito en el simulador
job = execute(qc, sim)

# Obtener el resultado
result = job.result()

# Obtener el estado vectorial del qubit
statevector = result.get_statevector(qc)

# Mostrar el vector de estado
print("State vector:", statevector)

# Visualizar en la esfera de Bloch
plot_bloch_multivector(statevector)
plt.show()
```
